### 5.21.7 Pengelompokan data menggunakan Pandas 

Data yang sama dapat dikelompokkan untuk analisis lebih lanjut.

```python
#Latihan mengelompokkan data menggunakan Pandas dengan perintah mean() (1)
bunga_iris.groupby(['sepal width (cm)']).mean()
#kode diatas menggunakan metode 'groupby()' dari pandas untuk engelompokkan baris di Data Frame berdasarkan nilai dari kolom 'sepal width(cm)'
#kemudian metode 'mean()' untuk menghitung nilai rata-rata dari setiap kelompok
#ketika kode diatas dijalankan, akan dihasilkan suatu data bru yang berisi nilai rata-rata dari setiap kolom 'bunga iris' yg sudah dikelompokkan berdasarkan nilai kolom 'sepal width (cm)'
```

```python
#Latihan mengelompokkan data menggunakan Pandas dengan perintah mean() (2)
bunga_iris.groupby(['sepal width (cm)']).mean().sort_values('sepal length (cm)', ascending=True)
#Menggunakan metode 'groupby()' untuk mengelompokkan baris dalam DataFrame berdasaekan nilai 'sepal width(cm)'
#Menggunakan metode 'mean()' untuk menghitung rata-rata dari setiap kelompok.
#Menggunakan metode 'sort_values()' untuk mengurutkan data hasil kelompokkan berdasarkan kolom 'sepal length (cm)' secara menaik
#ketika kode tersebut dijalankan, output yang dihasilkan adalah data baru yang berisi nilai rata-rata dar setiap kolom yang sudah dikelompokkan berdasarkan nilai pada kolom 'sepal width (cm)
#Kemudian, data diurutkan berdasarkan kolom 'sepal length(cm)' secara menaik
```

```python
#Latihan mengelompokkan data menggunakan Pandas dengan perintah sum (3)
bunga_iris.groupby(['sepal width (cm)']).sum()
#pada perintah diatas, dilakukan suatu pengelompokkan data pada variabel bunga_iris berdasarkan kolom 'sepal width (cm)'
#Selanjutnya dilakukan agregasi nilai dengan fungsi 'sum()' 
#dimana untuk setiap kelompok data yang dihasilkan dari pengelompokkan tersebut
#Output yang dihasilkan adalah suatu data hasil penjumlahan untuk setiap kolom dari data bunga_iris 
#dengan perhitungannya berdasarkan kelompok-kelompok yang dibentuk berdasarkan nilai unik pada kolom 'sepal width(cm)'
```

```python
#Latihan mengelompokkan data menggunakan pandas dengan perintah count (4)
#perintah grupby() digunakan untuk mengelompokkan data berdasaekan nilai unik pada suatu kolom yaitu 'sepal width (cm)'
#perintah count() digunakan untuk menghitung jumlah entri data pada setiap kelompok
bunga_iris.groupby(['sepal width (cm)']).count()
#kode diatas akan menghasilkan output berupa data yang menunjukkan jumlah entri data pada setiap nilai unik pada kolom 'sepal width (Cm)'
```

## 5.22 Matplotlib

Matplotlib merupakan salah satu Library pada Python yang fungsi utamanya
adalah dalam hal menggambarkan grafik, histogram dan berbagai jenis plot hasil
dari pengolahan data menggunakan Library lainnya yaitu Pandas dan Numpy.
Untuk itu penggunaan ketiga Library tersebut biasanya menjadi satu kesatuan.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

### 5.22.1 Grafik sederhana dengan Matplotlib 

Pada dasarnya Matplotlib berfungsi untuk memplot data menjadi grafik, histogram atau bentuk lainnya untuk mempermudah visualisasi data yang sedang
diolah.

```python
#Latihan mengenai Matplotlib dengan grafik sederhana (1)
#insialisasikan variabel x dan y yang merukan list dari number yang berisi 3 nilai di dalam masing-masing variabel
x = [2,4,6]
y = [5,8,10]

#digunakan metode plot yang akan menggambar garis yang menghubungkan ketiga titik yaitu (2,5), (4,8), dan (6,10) pada bidang koordinat
plt.plot (x,y)
#Kemudian, kode dibawah digunakan untuk memberikan judul grafik yaitu grafik y = x+2
plt.title ("Grafik y = x+2")
#kemudian, sumbu x diberi label dengan kode sebagai berikut
plt.xlabel ('x')
#begitu pula dengan label pada sumbu y
plt.ylabel ('y')
#output yang dihasilkan adalah sebuah grafik yang menunjukkan hubungan antara variabel x dan y
#pada grafik diberi judul 'Grafik y = x + 2', serta label dari sumbu x dan y
```

### 5.22.2 Properti pembuatan grafik dengan Matplotlib 

Untuk memperjelas keterangan pada grafik, ada banyak properti yang bisa dimanfaatkan pada Matplotlib.

```python
#Latihan properti pembuatan grafik dengan Matplotlib (1)
# diinisialisasikan variabel-variabel yang berisi data kolom pada dataFrame bunga_iris
x = bunga_iris['sepal length (cm)']
y = bunga_iris['petal length (cm)']
x1 = bunga_iris['sepal width (cm)']
y1 = bunga_iris['petal width (cm)']


plt.figure(figsize=(20,20), dpi=300)
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 8x8 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci



plt.plot(x,y, label = 'Grafik length' ,  color = 'yellow' , 
         linewidth = 3 , marker = '.' , markersize = 15 , 
         markeredgecolor = 'pink' , linestyle='-')
#label='Grafik length' : dugunakan untuk memberi label atau judul pada plot
#color : digunakan untuk memberi warna pada plot
#linewidth : digunakan untuk memberi ketebalan garis pada plot
#marker : digunakan untuk menandai titik pada plot dengan bentuk titik
#markersize : digunakan untuk menentukan ukuran dari marker
#markerredgecolor : digunakan untuk memberi warna pada tepi marker
#linestyle : digunakan untuk menentukan jenis garis pada plot


plt.plot(x1,y1, label = 'Grafik width', marker = '*' , 
         markersize = 15 , markeredgecolor = 'red' )
#metode ini digunakan untuk membuat plot dari data pada variabel x1 dan y1
#plot ini tidak memiliki variabel lain seperti label ataupun color sehingga plot ini akan menggunakan nilai default untuk setiap parameter

# berikut penamaan koordinat dengan fungsi 'title()'
# fungsi 'title()' digunakan untuk memasukan data string dan ukuran font
# fungsi 'fontdict' digunakan untuk menentukan ukuran font
plt.title('Grafik Bunga Iris ', fontdict = {'fontname' : 'Times New Roman' , 'fontsize' : 15})

# berikut inisialisasi sumbu x dan y dengan penamaan menggunakan data string
# fungsi 'fontdict' digunakan untuk menentukan ukuran font
plt.xlabel('X', fontdict={'fontname' : 'Comic Sans MS' , 'fontsize' : 6})
plt.ylabel('Y', fontdict={'fontname' : 'Comic Sans MS' , 'fontsize' : 6})

# berikut pembentukan output dari label x dan y
plt.xticks(bunga_iris['sepal length (cm)'])
plt.yticks(bunga_iris['petal length (cm)'])
plt.legend()

# berikut pembentukan output dari label x1 dan y1
plt.xticks(bunga_iris['sepal width (cm)'])
plt.yticks(bunga_iris['petal width (cm)'])
plt.legend()

# berikut penyimpanan figure pada 'grafik.png' dengan resolusi 500 dpi
plt.savefig('grafik.png' , dpi=500)

# berikut fungsi 'show()' yang digunakan untuk mengeluarkan output figure
plt.show()
```

### 5.22.4 Properti pembuatan histogram dengan Matplotlib 

Berbagai properti pembuatan histogram dengan menggunakan Matplotlib.

```python
#Latihan pembuatan histogram dengan Matplotlib (1)
#Diinisialisasikan variabel pekerjaan yang berisi list untuk menjadi label pada sumbu x
pekerjaan = ['Dokter', 'Dosen', 'Bisnis', 'Programmer', 'Petani']
#diinisialisasikan juga variabel jumlah sebagai nilai pada sumbu y
jumlah = [70, 52, 100, 37, 25]

#digunakan method "plt.figure" untuk menentukan ukuran dan resolusi gambar
plt.figure(figsize=(8,9))
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 8x9 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci

histogram = plt.bar(pekerjaan,jumlah,color = 'pink')
#Selanjutnya kita memanggil metode untuk membuat histogram
#Kita juga dapat menentukan warna pada histogram dengan menambahan parameter color
#Variabel histogram berisi sebuah list yang berisi bar chart yang telah dibuat.
pola = ['*','/','o','x','.']
#Kita juga dapat membuat sebuah list 'pola' yang berisi simbol-simbol yang digunakan sebagai pola pada bar chart

#kemudian, kita dapat mengatur pola pada setiap bar chart dengan menggunakan metode '.set.hatch()' pada masing-masing bar chart yang ada pada variabel'histogram'
histogram [0].set_hatch('*')
histogram [1].set_hatch('/')
histogram [2].set_hatch('o')
histogram [3].set_hatch('x')
histogram [4].set_hatch('.')
plt.show()
#kode diatas dipanggil untuk menampilkan histogram yang telah dibuat
```

## Program Menggunakan data iris dengan Pandas dan Matplotlib 

```python
#Latihan Program menggunakan data iris dengan Pandas dan Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#membaca file data iris
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                 names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])

#menampilkan lima data pertama dari dataframe
print(df.head())

#menghitung nilai rata-rata setiap variabel untuk setiap jenis bunga iris
df_mean = df.groupby(['class']).mean()
print(df_mean)

#membuat plot scatter untuk melihat persebaran bunga iris berdasarkan panjang dan lebar kelopak bunga
plt.scatter(df[df['class']=='Iris-setosa']['sepal length'], df[df['class']=='Iris-setosa']['sepal width'], 
            color='pink', label='Iris-setosa')
plt.scatter(df[df['class']=='Iris-versicolor']['sepal length'], df[df['class']=='Iris-versicolor']['sepal width'], 
            color='green', label='Iris-versicolor')
plt.scatter(df[df['class']=='Iris-virginica']['sepal length'], df[df['class']=='Iris-virginica']['sepal width'], 
            color='blue', label='Iris-virginica')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter plot of Iris dataset')
plt.legend()
plt.show()
```
